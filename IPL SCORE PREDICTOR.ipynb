```python
new_df=pd.merge(df,df2,on="ID")
new_df.head(3)
```

```python
new_df.info()
```

```python
req_df=new_df[['ID', 'innings', 'overs', 'ballnumber', 
          'total_run',
       'non_boundary', 'isWicketDelivery',  
       'BattingTeam', 'City',  
        'Team1', 'Team2', 'Venue', 'TossWinner', 'TossDecision']]
req_df.head()
```

```python
req_df=req_df[req_df['innings']==1]
req_df[req_df['BattingTeam']==req_df['Team1']]
```

```python
req_df.shape
```

```python
req_df['BowlingTeam']=np.where(req_df['Team2']==req_df['BattingTeam'],req_df['Team1'],req_df['Team2'])
```

```python
req_df.sample(5)
```

```python
req_df.columns
```

```python
req_df=req_df[['ID',  'BattingTeam', 
        'BowlingTeam','overs', 'ballnumber', 'total_run', 
       'isWicketDelivery', 'City']]
```
